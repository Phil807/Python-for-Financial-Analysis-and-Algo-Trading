3 ways to plot a graph in pandas visualization

```python
df1['A'].hist(bins=30)   #notice it is calling matplotlib under the hood
                         #i can add in matplotlib or even seaborn style arguments
```

```python
df1['A'].plot(kind='hist',bins=30)
```

```python
df1['A'].plot.hist(bins=30) #this way is useful as we can call the different types of plot of the .plot method
```

```python
df2.head()
```

```python
df2.plot.area(alpha=0.5) #its a line plot with the area underneath the line filled in with color
```

```python
df2.plot.bar() #x axis contain the index
               #y axis is the df values
               #legend shows the df columns
```

```python
df2.plot.bar(stacked=True)
```
