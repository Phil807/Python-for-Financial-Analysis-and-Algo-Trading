```python
plt.subplot(1,2,1)  #rows, columns, plot number I am referring to
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OBJECT ORIENTATED MATPLOTLIB METHOD
#create figure objects and then call methods off the figure objects
```

```python
fig = plt.figure() #using the figure() function from pyplot library to create a figure object
                   #the figure object is like a blank canvas

    
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #called the figure method add_axes to create an axes object 

axes.plot(x,y)                         #we then used the plot method to plot on the axes
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('First Object Orientated Plot')
```

```python
fig = plt.figure()   #using the figure() function from the pyplot library to create a figure object

    
axes1 = fig.add_axes([0,0,1,1])         #the list that I pass in to add_axes method will now make sense 
                                        #when we have two plots
axes2 = fig.add_axes([0.2,0.5,0.4,0.4]) #20% in from the left corner of the emply figure canvas, 50% up from the bottom, 
                                        #40% of original canvas in width, 40% of original canvas in height

    

axes1.plot(x,y)  #notice here plot is a method called off the axes object
axes1.set_title('Larget Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

object oriented method

1. Create a figure object (this is essentially just a blank canvas)
2. Add a set of axes to the figure (pass in a list to determine the axes location in relation to the blank canvas figure)
3. Plot a graph
4. Add extras such as plot title if you wish

Matplotlib Basics Part 2

```python
#creating subplots using OOP method
```

```python
fig, axes = plt.subplots() #create the figure and axes in one line of code

axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2) #subplots() function creates the figure and addess the axes in one go
```

```python
fig, axes = plt.subplots(nrows=4,ncols=4)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

for current_axes in axes:
    current_axes.plot(x,y)
```

```python
axes  #the axes object contains an array of matplotlib axes objects
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

#as we can iterate through an axes so we can also index off of an axes

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()  
```

object oriented subplots approach (cuts out a step)

1. Create a figure and axes using subplots function
3. Index the axes and plot a graph
4. Add extras such as plot title if you wish

```python
#FIGURE SIZE AND DPI(DOTS PER INCH)
```

```python
fig = plt.figure(figsize=(3,2))

axe = fig.add_axes([0,0,1,1])
axe.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(3,2)) #nrows=1 and ncols=1 by default

axes.plot(x,y) #notice I dont need to index the axes if only one axes available to choose
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(5,2)) 

axes[0].plot(x,y) 

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) #this adds the firsts x and y axis
axes1.plot(x,y,label='X Squared') #must add in a label for the legend() method to work
axes1.plot(x,y+4, label ='Y + 4')  

axes1.legend()
```

Matplotlib Part 3

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='green', linewidth=2, alpha=0.5, linestyle='--', marker='o', markersize = 5,
       markerfacecolor='black', markeredgewidth=4, markeredgecolor = 'blue')  

#alpha controls line transparancy
#marker marks out each x,y co-ordinate point
```
