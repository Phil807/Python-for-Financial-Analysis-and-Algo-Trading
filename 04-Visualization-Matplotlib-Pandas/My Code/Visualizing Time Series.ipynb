```python
import matplotlib.dates as dates  #can also write from matplotlib import dates
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01','2009-01-01'], ylim=(0,50)) 
```

```python
idx = mcdon.loc['2007-01-01':'2007-05-01'].index 
```

```python
idx
```

```python
stock = mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
```

```python
stock.head()
```

```python
#now lets plot on matplotlib
```

```python
fig, ax = plt.subplots()

ax.plot_date(idx,stock,'-')        #plot_date method called off ax lets matplotlib know that we 
ax.yaxis.grid(True)                 #are dealing with a datetime index array
ax.xaxis.grid(True)                 #this lets matplotlib know we can call things like year, month, week and day off of it

fig.autofmt_xdate()                 #automatically formats the x date axis

#LOCATING AND FORMATTING THE AXIS

#Locating
ax.xaxis.set_major_locator(dates.MonthLocator())  #locates the month tick on the x axis

#Formatting
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n%b--%Y'))  



#CAN ALSO INTRODUCE A MINOR X AXIS

#Locating
ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0)) #byweekday chooses what weekday to locate, 0 is Monday

#Formatting
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))  
```
