```python
df = pd.read_csv('time_data\walmart_stock.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
df.rolling(window=7).mean().head(14)

#notice the first 6 rows are Nan are I am requesting a mean based off 7 rows of data
#each mean value is calculated based off the previous 7 days
#so the 12th uses open values from the 4th to the 12th to calculate the mean
#one day drops off and another is added in
#this round out a lot of noise, big jumps.
```

```python
df['Open'].plot()
df.rolling(window=21).mean()['Open'].plot(figsize=(16,5))  #apply rolling average to the whole dataframe and plot Open data
```

```python
df['Open'].plot()
df['Open'].rolling(window=21).mean().plot(figsize=(16,5)) #apply rolling average to Open column and plot Open data
```

```python
#the longer the window the more you will see the long term trend, it will filter out daily noise
```

```python
df['Close 12 Day Moving Average'] = df['Close'].rolling(window=12).mean()
```

```python
df.head(15)
```

```python
df[['Close','Close 12 Day Moving Average']].plot(figsize=(16,6))#notice a legend gets added in when I make the Moving 
                                                                #Average a column in my df
```

```python
#rolling - takes for example the previous 7 days to calculate a value. One figure gets added in and another drops out
#expanding - figures only get added in, nothing drops out
```

```python
df['Close'].expanding().mean().plot(figsize=(16,5))
```

```python
#so the above chart shows us that since the beginning of the time series out price has rises
#it wont reflect something like a small drop in the middle of the time series if all the values before it are positive 
```

```python
#BOLLINGER BANDS
```

```python
#bands places above and below the moving average where the volatility is based on std which changes as volatility
#increases or decreases. The bands will widen as volatility increases and tighten as volatility decreases
```

```python
#can be used to identify tops and bottoms or to determine the strength of the actual trend
#bollinger bands tend to use a 20 day period as their moving average
#prices are relatively high when the are above the upper bands and vice versa
```

```python
#4 Steps

#Create Close 20MA Column
df['Close 20 Day Moving Average'] = df['Close'].rolling(window=20).mean()

#Upper Band = 20MA + 2*std(20)
df['Upper'] = df['Close 20 Day Moving Average'] + 2*(df['Close'].rolling(20).std())

#Lower Band = 20MA - 2*std(20)
df['Lower'] = df['Close 20 Day Moving Average'] - 2*df['Close'].rolling(20).std()

#Plot the above 3 columns and the original Close column
df[['Close', 'Close 20 Day Moving Average', 'Upper', 'Lower']].plot(figsize=(16,6))
```

```python
#2*(df['Close'].rolling(20).std()   this is basically our volatility. Our rolling std over the past 20 days * 2
#2 std's avway from the moving average
```
