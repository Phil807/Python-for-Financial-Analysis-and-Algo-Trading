```python
df = pd.read_csv('time_data\walmart_stock.csv',index_col='Date',parse_dates=True)
```

```python
df.shift(periods=1).head() #all the index dates get shifted up 1 row. The 4th now takes the 3rd values. 
                           #the 3rd has been shifted up one row to where there is no data
                           #we also lose the very last row of data as the last date gets shifted up 1 row and so the last 
                           #row of data now has no index value so it drops out of the df
```

```python
#shift backwards by setting periods to a negative value

df.shift(periods=-1).head()
```

```python
#tshift method and frequency argument
#shifts all the rows to the matching time string code that you provide
```

```python
df.head()
```

```python
df.tshift(freq='M').head()  #gives everything in each month the same time stamp
                            #note tshift and shift are the same thing
```

```python
df.tshift(freq='A').head() #gives every date in the same year the same datetime
```
