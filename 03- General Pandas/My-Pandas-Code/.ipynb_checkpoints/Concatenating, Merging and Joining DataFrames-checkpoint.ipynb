```python
#concatenating glues all the info in seperate tables together
#merge works on columns
#join works on rows using the index as the key(s)
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
#Concatenation basically glues together DataFrames. 
#Keep in mind that dimensions should match along the axis you are concatenating on.Default axis = 1 
#You can use pd.concat and pass in a list of DataFrames to concatenate together
```

```python
pd.concat([df1,df2,df3])  #by default it concatenates all the rows
```

```python
pd.concat([df1,df2,df3], axis = 1) #which axis you concatenate on depends on what makes most sense for the given data
```

```python
#Merging

#merge is a function taken from the pd library
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})   
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how = 'inner', on = 'key') 

#left and right are the dataframes and the order I want them shown in the output
#how = 'inner' by default. It means keep the order of the keys taken from the left dataframe
#on = 'key means join on the key column
```

```python
#more examples
```

```python
dframe1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
dframe2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
dframe1
```

```python
dframe2
```

```python
#can merge using more than one key.
#how = 'inner' means both keys need to be in both tables and match for the row to merge
```

```python
pd.merge(dframe1,dframe2,on = ['key1','key2'])
```

```python
#can inner lead to some data being counted more than you want? See A2 B2 in the merged table.
```

```python
pd.merge(dframe1,dframe2,how = 'outer',on = ['key1','key2'])
#how = 'outer' means the output returns all key pairs
#left dataframe keys take the hierarchial order
```

```python
pd.merge(dframe1,dframe2,how = 'right',on = ['key1','key2'])
#how = 'right' returns all the keys from the right dataframe 
```

```python
pd.merge(dframe1,dframe2,how = 'left',on = ['key1','key2'])
#takes all the keys from the left table. Notice left dataframe data A2 B2 is returned twice in the output
```

```python
#joining
#Joining is a convenient method for combining the columns of two potentially 
#differently-indexed DataFrames into a single result DataFrame.
#same as merge but instead the keys are on the index instead of the column
```

```python
dleft = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

dright = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
dleft
```

```python
dright
```

```python
#join is a method of a dataframe
```

```python
dleft.join(dright) #default is a left join. Takes all the index keys from the dleft dataframe
```

```python
dleft.join(dright, how = 'outer') #how = 'right' returns all index keys
```

```python
#inner - if keys match in both tables
#outer - all keys regardless
#left - keys in the left table
#right - keys in the right table
```
