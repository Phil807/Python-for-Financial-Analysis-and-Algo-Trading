```python
bank_lst = list(bankdf['Bank Name'])
```

```python
def bank_name(bank_lst):
    
    count = 0
    
    for name in bank_lst:
        if 'Bank' not in name:
            count += 1
            
    return count
```

```python
bank_name(bank_lst)
```
