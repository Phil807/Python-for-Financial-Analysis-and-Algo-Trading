```python
#ARIMA - Auto Regressive Integretated Moving Average
#ARIMA do not work well with time series financial data however they work well with other data such as trying to predict
#the volume of sales a store will do in the summer etc
```

```python
#ARIMA is a generalized form of the ARMA model
#ARIMA model can be used to better understand time series data and to predict future points in the series (forecasting)
```

```python
#2 types
#non seasonal ARIMA models
#seasonal ARIMA models (more complex)
```

```python
#3 components to ARIMA models - 1. Autoregressive 2. Integrated 3. Moving Average
```

```python
#non seasonal ARIMA models are generally denoted ARIMA(p,d,q) where p,d and q are non-negative integers
```

```python
#AR(p): Autoregressive
 #A regression model that utilizes the dependent relationship between a current observation and 
 #observations over a previous period
```

```python
#I(d): Integrated
 #Differencing of observations (subtracting an observation from an observation at the previous time step) in order
 #to make the time series stationary
```

```python
#MA(q): Moving Average
 #A model that uses the dependency between an observation and a residual error from a moving average model
 #applied to lagged observations
```

```python
#stationary vs non stationary data
```

```python
#a stationary dataset has constant mean and variance over time
#a stationary dataset will allow our model to predict that the mean and variance will be the same in future periods 
```

```python
#a common test to test for a datasets mean, variance and covariance over time is the Dickey-Fuller test
#we can do this test using the statsmodels library
```

```python
#if your dataset is non-stationary you will need to transform it to a stationary dataset in order to use
#an ARIMA model on it. One simple way to convert your dataset to a stationary dataset is to use 'Differencing'.
#see lecture 69, around 7 minutes 40 seconds in for diffencing example.
```

```python
#You can continue differencing until you reach stationarity(which you can check visually and mathematically)
#note each differencing step will result in losing a row of data which is not a huge cost if you have a large dataset
#you can also do seasonal differencing
#can also take the seasonal difference of the first difference
```

```python
#steps
#1. check if data is stationary
#2. if not then can use differening technique to transform the data
#3. Apply the 3 parts of the ARIMA model
```

```python
#With autocorrelation in general you would use either AR or MA part of ARIMA.
#Using both is uncommon
```

```python
#if the autocorrelation plot shows positive correlation at the first lag (lag 1), then
#it suggests to use the AR terms in relation to the lag
#if the autocorrelation plot shows negative correlation at the first lag (lag 1), then
#it suggests to use the MA terms in relation to the lag
#this will help you to decide on what values to use for p,d,q for your ARIMA model
```

```python
#p - is usually decided by the number of lag observations included in the model
#d - the number of times the raw observations are different
#q - the size of the moving average window, also called the order of the MA
```

```python
#there are also Partial Correlation Plots 
#a conditional correlation
#it is a corrlelation between 2 variables under the assumption that we know and take into account the values of 
#some other set of variables
```

```python
#Partial Correlatoin example
#consider a regression in which y=response variable and x1, x2 and x3 are the predictor variables
#the partial correlation between y and x3 is determined taking into account and both y and x3 are related to x1 and x2
```

```python
#typicall a sharp drop off after lag 'k' suggests an AR-k model should be used
#a graducal decline suggests an MA models
```

```python
#identification of the AR model is best done with the PACF (Partial Auto Correlation Function)
#identification of the MA model is often best done with the ACF
```

```python
#finally once you have analyzed your data using ACF or PACF you are ready to begin to apply ARIMA or Seasonal ARIMA
#you will provide the p,d,q terms for the model (ARIMA model)
#seasonal ARIMA also requires additional P,D,Q terms
```

```python
#Step 3 - test if data is stationary 
```

```python
#use Dicky Fuller test
#hypothesis is that the data is non stationary
#alternative - series has no unit root and is stationary
```

```python
#samll p-value, less than 0.05 indicates strong evidence against the null
#0.05% chance the alternative occured due to chance
#can then reject the null

#large p-value indicates weak evidence against the null
#then we fail to reject the null
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(df['Milk in Pounds per Cow'])
```

```python
result
```

```python
def adf_check(time_series):
    
    result = adfuller(time_series)
    print('Augmented Dicky Fuller Test\n')
    labels = ['ADF Test Statistic', 'p-value', '# of lags','# of observations']
    
    for value,label in zip(result,labels):
        print(label+ " : " +str(value))
        
    if result[1] <= 0.05:
        print('\nStrong evidence against null')
        print('reject null')
        print('Data has no unit root and is stationary')
    else:
        print('\nWeak evidence againsts the null')
        print('Fail to reject the null')
        print('Data has a unit root and is non stationary')
```

```python
test_value = (4,5,6,3,6,9)
test_header = ['First','Second','Third']
list(zip(test_value,test_header))
```

```python
adf_check(time_series)
```

```python
#so the stationary is non stationary so we should find the difference to make it stationary
```

```python
df['First Difference'] = df['Milk in Pounds per Cow']-df['Milk in Pounds per Cow'].shift(1)
```

```python
df['First Difference'].plot(figsize=(10,4))
```

```python
adf_check(df['First Difference'].dropna()) #so now we have stationary data
```

```python
#lets just imagine we need to do a second difference to make the data stationary, for practice

df['Second Difference'] = df['First Difference']-df['First Difference'].shift(1)
```

```python
adf_check(df['Second Difference'].dropna())
```

```python
#data was seasonal so we use the seasonal ARIMA model
```

```python
from statsmodels.tsa.arima_model import ARIMA  #ARIMA model for non seasonal data
```

```python
#help(ARIMA)
```

```python
#Call the seasonal ARIMA model
```

```python
#assign the model
model = sm.tsa.statespace.SARIMAX(df['Milk in Pounds per Cow'],order=(0,1,0), seasonal_order=(1,1,1,12))
```

```python
#fit the model
results = model.fit()
```

```python
print(results.summary())
```

```python
results.resid.head() #gives the residual values
                     #the residual is the real value - the forecast value
```

```python
results.resid.plot(kind='kde')
```

```python
#create a forecast and compare to results we already have
df['Forecast'] = results.predict(start=150, end=168) 
df[['Milk in Pounds per Cow','Forecast']].plot()
```
