```python
#ETS - Error, Trends, Seasonality
#it is a good idea to plot the data of these 3 terms in order to get an understanding of your dataset
#the error term explains the movement in the data that is not explained but the trend term or the seasonality term
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
airline = pd.read_csv('airline_passengers.csv',index_col='Month')
```

```python
airline.head()
```

```python
airline.plot()
```

```python
airline.dropna(inplace=True)
```

```python
airline.index = pd.to_datetime(airline.index)
```

```python
airline.index
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose #import the seasonal_decompose function
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'],model='multiplicative')
```

```python
result
```

```python
result.seasonal.plot() #now from the result variable we can call the seasonal data (or a bunch of other data)
```

```python
result.trend.plot()
```

```python
result.plot() #notice we get the same plots twice, this is just a bug in jupyter notebooks
              #to avoid this assign the output to a new variable as done in the next cell
```
